### Завдання 9.1
Створіть функцію `run_etl(source_path, dest_path)`, яка:
1. **Extract**: читає CSV (транзакції)
2. **Transform**:
   - видаляє рядки з пропущеними `transaction_id`
   - конвертує `date` в datetime
   - фільтрує тільки `status == 'completed'`
   - створює колонку `revenue = quantity * price`
   - групує по `product_category`: сума revenue, кількість транзакцій
3. **Load**: зберігає в Parquet

Застосуйте функцію до згенерованих тестових даних (1000 рядків).


```python
# Створіть тестові дані
def generate_transactions(n=1000):
    categories = ['Electronics', 'Clothing', 'Home', 'Books', 'Sports']
    statuses = ['completed', 'pending', 'failed', 'cancelled']
    data = {
        'transaction_id': [f'TXN{i:05d}' for i in range(n)],
        'date': [datetime(2024, 1, 1) + timedelta(days=random.randint(0, 364)) for _ in range(n)],
        'product': [random.choice(['Widget', 'Gadget', 'Doohickey', 'Thingy', 'Whatsit']) for _ in range(n)],
        'product_category': [random.choice(categories) for _ in range(n)],
        'quantity': [random.randint(1, 100) for _ in range(n)],
        'price': [round(random.uniform(1.0, 500.0), 2) for _ in range(n)],
        'status': [random.choice(statuses) for _ in range(n)],
        'customer_id': [random.randint(1000, 9999) for _ in range(n)]
    }
    df = pd.DataFrame(data)
    # Вносимо кілька пропусків
    df.loc[random.sample(range(n), 5), 'transaction_id'] = None
    df.loc[random.sample(range(n), 3), 'date'] = None
    return df

df_txn = generate_transactions(1000)
os.makedirs('output', exist_ok=True)
df_txn.to_csv('output/transactions.csv', index=False)
print("Test data created: output/transactions.csv")
print(df_txn.head(3))
```
